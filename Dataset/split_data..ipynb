```python
import pandas as pd
import os

file_path = 'E:\C\ICT\Machine Learning and Data Mining\ML-DM-Stock-Price-Prediction\Dataset\stock_data_ml.csv'

df = pd.read_csv(file_path)

tickers = df['ticker'].unique()

base_folder = os.path.dirname(file_path)

for ticker in tickers:
    ticker_df = df[df['ticker'] == ticker].reset_index(drop=True)
    ticker_df.index += 1  
    ticker_df.to_csv(os.path.join(base_folder, f"{ticker}.csv"), index=True)

print("CSV files have been split and saved successfully.")
```

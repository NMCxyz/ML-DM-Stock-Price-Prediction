```python
# Initation the dataframe
data = pd.DataFrame()
```

```python
# Declare Funtion to get the data of the ticker
def getDataTicker(ticker):
    end_date = date.today().strftime('%Y-%m-%d')
    try:
        df = stock_historical_data(symbol=ticker, 
                                    start_date="2024-01-01", 
                                    end_date=end_date, resolution='1D', type='stock', beautify=True, decor=False, source='DNSE')
    except:
        df = pd.DataFrame()
    return  df
```

```python
# List of the ticker we use in this project
list = {'TV2', 'VCS', 'HPG', 'VNM', 'REE', 'FPT', 'VIC', 'MWG', 'VCB'}
```

```python
# Crawling the data
for ticker in list:
    data = pd.concat([getDataTicker(ticker), pd.DataFrame(data)], ignore_index=True)
```

```python
#Save the data to csv file
data.reset_index()
data.to_csv('stock_data_ml_extend.csv')
```
